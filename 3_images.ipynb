## Images and filters

2D FT for image data processing.

### 3.1 Build image

Create 256x256 matrix filled with zeros and insert

$$\text{image1}(i,j) = \sin{(\frac{2 \pi (i + j)}{5})}$$

```python
# Load packages
import numpy as np
import matplotlib.pyplot as plt

# create zeros matrix as an image
im0 = np.zeros((256, 256), dtype=np.float32)
nrow, ncol = im0.shape # get the shape of the image

# create a sinusoidal image
for i in range(nrow):
    for j in range(ncol):
        im0[i, j] = np.sin(2 * np.pi * (i + j) / 5)

# plot the image
plt.imshow(im0, cmap='gray')
```

### 3.2 Calculate the 2D FT and plot

Calculate the **2D FT** of this image using the `fft2` function, and draw the amplitude and phase spectra. We will use the `fftshift` function to put the zero frequency in the center of the image. Interpret the spectra with your knowledge of the Fourier transform.

```python
# FT in 2D
ff0=np.fft.fftshift(np.fft.fft2(np.fft.fftshift(im0)))   # First center the image, then FT, then center FT
# arrange that in subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(im0, cmap='gray')
ax[0].set_title('Original image')
ax[1].imshow(np.abs(ff0), cmap='gray')
ax[1].set_title('Fourier transform')
ax[2].imshow(np.angle(ff0), cmap='gray')
ax[2].set_title('Phase')
```

### 3.3 Same with Gaussian distribution

Perform the same analysis with a **Gaussian** function centered on `𝑂 = (129, 129)`, which is the middle of the image.

```python
# Perform the same with a Gaussian centered in the middle
# create zeros matrix as an image
im1 = np.zeros((256, 256), dtype=np.float32)
nrow, ncol = im1.shape # get the shape of the image

# create a gaussian image
# Define center
nx = nrow / 2
ny = ncol / 2
for i in range(nrow):
    for j in range(ncol):
        im1[i, j] = np.exp(-((i - nx)**2 + (j - ny)**2) / (2 * np.pi ** 2))

# FT in 2D
ff1=np.fft.fftshift(np.fft.fft2(np.fft.fftshift(im1)))
# arrange that in subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(im1, cmap='gray')
ax[0].set_title('Original image')
ax[1].imshow(np.abs(ff1), cmap='gray')
ax[1].set_title('Fourier transform')
ax[2].imshow(np.angle(ff1), cmap='gray')
ax[2].set_title('Phase')
```

A 2D **rectagle** function centered on `𝑂`.

```python
# Perform the same with a Rectangle centered in the middle
# create zeros matrix as an image
im3 = np.zeros((256, 256), dtype=np.float32)
nrow, ncol = im3.shape # get the shape of the image

# create a rectangle image
# Define center
nx = nrow / 2
ny = ncol / 2

# Define width and height
w = 50
h = 20
for i in range(nrow):
    for j in range(ncol):
        if (i > nx - w / 2) and (i < nx + w / 2) and (j > ny - h / 2) and (j < ny + h / 2):
            im3[i, j] = 1

# FT in 2D
ff1=np.fft.fftshift(np.fft.fft2(np.fft.fftshift(im3)))
# arrange that in subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(im3, cmap='gray')
ax[0].set_title('Original image')
ax[1].imshow(np.abs(ff1), cmap='gray')
ax[1].set_title('Fourier transform')
ax[2].imshow(np.angle(ff1), cmap='gray')
ax[2].set_title('Phase')
```
